### Señal continua y discreta

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Señal continua
tiempo = np.arange(-60, 60, 0.01) # Tiempo continuo 
signal = 3*np.cos(0.1*np.pi*tiempo) # Señal análoga 
```

```python
#Señal discreta
tiempod = np.arange(-60, 60, 1) # Tiempo discreto
signald = 3*np.cos(0.1*np.pi*tiempod) #Señal discreta
```

```python
#Gráficas
plt.figure(figsize=(20, 12))

#Gráfica señal continua
plt.subplot(2, 2, 1)
plt.title('Señal continua',fontsize=14)
plt.xlabel('Tiempo ',fontsize=14)
plt.ylabel('Amplitud [Voltios]',fontsize=14)
plt.plot(tiempo,signal, color = 'm')
plt.grid()

#Gráfica señal discreta
plt.subplot(2, 2, 3)
plt.title('Señal discreta',fontsize=14)
plt.xlabel('Tiempo',fontsize=14)
plt.ylabel('Amplitud [Voltios]',fontsize=14)
plt.stem(tiempod,signald, linefmt = 'g', markerfmt = 'go' )
plt.grid()

#Gráfica contina y discreta
plt.subplot(2, 2, 2)
plt.title('Señal continua y discreta',fontsize=14)
plt.xlabel('Tiempo',fontsize=14)
plt.ylabel('Amplitud [Voltios]',fontsize=14)
plt.stem(tiempod,signald, label= 'Señal discreta',linefmt = 'g', markerfmt = 'go' )
plt.plot(tiempo,signal, label = 'Señal continua',color = 'm')
plt.legend()
plt.grid()

plt.show()
```
