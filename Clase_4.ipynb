## Librerias

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

### Suma

```python
sec1 = np.random.randint(0,20,15)
print('Secuencia 1: ',sec1)

sec2 = np.random.randint(0,20,15)
print('Secuencia 2: ',sec2)

suma = sec1+sec2
tiempo = np.arange(len(suma)) 

plt.figure(figsize=(18,15))
plt.subplot(3,1,1)
plt.title('Secuencia 1')
plt.xlabel('Muestras')
plt.ylabel('Amplitud')
plt.stem(tiempo,sec1)
plt.grid()

plt.subplot(3,1,2)
plt.title('Secuencia 2')
plt.xlabel('Muestras')
plt.ylabel('Amplitud')
plt.stem(tiempo,sec2)
plt.grid()

plt.subplot(3,1,3)
plt.title('Suma de secuencia')
plt.xlabel('Muestras')
plt.ylabel('Amplitud')
plt.stem(tiempo,suma)
plt.grid()

plt.show()
```

### Multiplicacin

```python
multi = sec1*sec2
print('Secuencia 1: ',sec1)
print('Secuencia 2: ',sec2)

tiempo = np.arange(len(multi)) 

plt.figure(figsize=(18,15))
plt.subplot(3,1,1)
plt.title('Secuencia 1')
plt.xlabel('Muestras')
plt.ylabel('Amplitud')
plt.stem(tiempo,sec1)
plt.grid()

plt.subplot(3,1,2)
plt.title('Secuencia')
plt.xlabel('Muestras')
plt.ylabel('Amplitud')
plt.stem(tiempo,sec2)
plt.grid()

plt.subplot(3,1,3)
plt.title('Multiplicacin de secuencia')
plt.xlabel('Muestras')
plt.ylabel('Amplitud')
plt.stem(tiempo,multi)
plt.grid()

plt.show
```

### Desplazamiento de seales Periodicas



```python
signal = [4, 6, 8, 10, 12, 14,16,18]
print('Secuencia para seal: ',signal)
t = np.arange(len(signal))

sigdes = []

desplaza = 2

cont = 0
for j in range(desplaza):
    sigdes.append(signal[-(j+desplaza)+ cont])
    cont += desplaza
for i in range(len(signal)):    
    sigdes.append(signal[i])
print('Secuencia de seal desplazada: ',sigdes)

tiempo = np.arange(len(sigdes)) 

#Grficas
plt.figure(figsize=(15,10))
plt.subplot(2,1,1)
plt.title('Seal')
plt.xlabel('Muestras')
plt.ylabel('Amplitud')
plt.stem(t,signal)
plt.grid()

plt.subplot(2,1,2)
plt.title('Seal con desplazamiento')
plt.xlabel('Muestras')
plt.ylabel('Amplitud')
plt.stem(tiempo,sigdes)
plt.grid()
```
