### Convoluci√≥n

### Librerias 

```python
import numpy as np 
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
sec = np.random.randint(0,20,15)
print('Secuencia 1: ',sec)

h = np.random.randint(0,20,15)
print('Secuencia 2: ',h)

convol = np.convolve(sec,h,mode='full')
tiempo = np.arange(0,29) 

plt.figure(figsize=(15,5))
plt.title('Convoluci√≥n')
plt.xlabel('Muestras')
plt.ylabel('Amplitud')
plt.stem(tiempo,convol)
plt.grid()
plt.show()
```

**La muestra cero esta en la posici√≥n cero en ambas secuencias**
