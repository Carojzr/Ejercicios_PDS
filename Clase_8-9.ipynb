### Diagrama de polos y ceros a partir de la función de transferencia


## Librerias

```python
import matplotlib.pyplot as plt
import matplotlib.patches as paches

from scipy.signal import tf2zpk

%matplotlib inline
```

## Función de transferencia
$H[z] = \frac{z^2+2}{z^2-10z+8}$

```python
num = [1,0,2]
den = [1,-10,8]
```

```python
z,p,k = tf2zpk(num,den)
unit_c = plt.Circle((0,0),1, fill=False, color='darkcyan', ls='--')

fig, ax = plt.subplots()
ax.add_patch(unit_c)
plt.plot(p.real,p.imag,'x')
plt.plot(z.real,z.imag,'o')
plt.show()
```
