## Librerias

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
secuencia = np.random.randint(0,20,15)
print(secuencia)
```

```python
tiempo = np.arange(len(secuencia)) 
plt.figure(figsize=(15,5))
plt.title('Secuencia')
plt.xlabel('Muestras')
plt.ylabel('Amplitud')
plt.stem(tiempo,secuencia,)
plt.grid()
```
