### Submuestreo

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Se crea la señal
tiempo = np.arange(-60, 60, 0.01) # Tiempo continuo 
signal = 3*np.cos(1*np.pi*tiempo) # Señal análoga 

plt.figure(figsize=(15,5))
plt.title('Señal continua')
plt.xlabel('Tiempo')
plt.ylabel('Amplitud [Voltios]')
plt.plot(tiempo,signal, color = 'm')
plt.grid()
```

```python
#Submuestreo
fmax=0.5
fm=0.2

sgnald = np.arange(-60, 60, 1.0/fm)
signal = 3*np.cos(1*np.pi*sgnald)
plt.figure(figsize=(15,5))
plt.title('Señal submuestreada')
plt.xlabel('Tiempo')
plt.ylabel('Amplitud [Voltios]')
plt.plot(sgnald,signal, color = 'm')
```
